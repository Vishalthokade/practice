```python
#explore data by attrition
from matplotlib import style
from matplotlib import gridspec
import matplotlib.pyplot as plt
import numpy as np
style.use('ggplot')
fig = plt.figure()
 
# to change size of subplot's
# set height of each subplot as 8
fig.set_figheight(18)
 
# set width of each subplot as 8
fig.set_figwidth(20)
plt.subplot(3,3,1)
plt.hist(leftempdata['Age'],color='red', edgecolor='black')
plt.title('Age wise distribution of left employees')
plt.ylabel('# of employees')
plt.xlabel('Age\n')
#print('\n')
plt.subplot(3,3,2)
plt.hist(leftempdata['DistanceFromHome'],color='yellow', edgecolor='black')
plt.title(' Distance From Home of left employees')
plt.ylabel('# of employees')
plt.xlabel('distance(in kms)')
#print('\n')
plt.subplot(3,3,3)
plt.hist(leftempdata['Education'],color='orange', edgecolor='black')
plt.title('Education of left employees')
plt.ylabel('# of employees')
plt.xlabel('Education years')

plt.subplot(3,3,4)
plt.hist(leftempdata['NumCompaniesWorked'],color='pink', edgecolor='black')
plt.title('NumCompaniesWorked of left employees')
plt.ylabel('# of employees')
plt.xlabel('# of Companies worked in')

plt.subplot(3,3,5)
plt.hist(leftempdata['MaritalStatus'],color='green', edgecolor='black')
plt.title('Marital Status of left employees')
plt.ylabel('# of employees')
plt.xlabel('Marital Status')

plt.subplot(3,3,6)
plt.hist(leftempdata['EducationField'],color='cyan', edgecolor='black')
plt.title('EducationField of left employees')
plt.ylabel('# of employees')
plt.xlabel('EducationField of left Employees')

plt.subplot(3,3,7)
plt.hist(leftempdata['EnvironmentSatisfaction'],color='black', edgecolor='black')
plt.title('Environment Satisfaction of left employees')
plt.ylabel('# of employees')
plt.xlabel('Environment Satisfaction of Left employees ')

plt.subplot(3,3,8)
plt.hist(leftempdata['JobSatisfaction'],color='blue', edgecolor='black')
plt.title('Job Satisfaction of left employees')
plt.ylabel('# of employees')
plt.xlabel('Job Satisfaction of left employees')

plt.subplot(3,3,9)
plt.hist(leftempdata['WorkLifeBalance'],color='brown', edgecolor='black')
plt.title('Work Life Balance of left employees')
plt.ylabel('# of employees')
plt.xlabel('Work Life Balance of left employees')
plt.tick_params(axis='x', labelrotation = 0)
plt.show()
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```
